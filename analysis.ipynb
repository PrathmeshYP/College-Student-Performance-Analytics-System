```python
df.drop_duplicates(inplace=True)
```

```python
df['Gender']=df['Gender'].map({'Male':0,'Female':1})
df['Final_Result']=df['Final_Result'].map({'Fail':0,'Pass':1})
```

```python
df["Total_Score"] = (
    df["Internal_Marks"] +
    df["Assignment_Score"] +
    df["Final_Marks"]
)
df["Study_Efficiency"] = df["Final_Marks"] / df["Study_Hours"]
```
